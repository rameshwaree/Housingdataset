```python
import pandas as pd #getting the appropriate packages
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
data = pd.read_csv('/content/housing (1).csv') #reading the data
```

```python
# describing the data
data.describe()
```

```python
#regression line and scatter plot between the chosen 2 varibale
plt.figure(figsize=(10, 6))
plt.scatter(data['housing_median_age'], data['median_income'], color='green', label='Data points')
plt.xlabel('Housing Median Age')
plt.ylabel('Median Income')
plt.title('Scatter plot and regression line between Housing Median Age and Median Income')

# Fit regression line
model = LinearRegression()
model.fit(data[['housing_median_age']], data['median_income'])
plt.plot(data['housing_median_age'], model.predict(data[['housing_median_age']]), color='red', label='Regression line')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['median_income'], data['median_house_value'], color='green', label='Data points')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.title('Scatter plot and regression line between Median Income and Median House Value')

# Fit regression line
model.fit(data[['median_income']], data['median_house_value'])
plt.plot(data['median_income'], model.predict(data[['median_income']]), color='red', label='Regression line')

plt.legend()
plt.show()
```

```python
#splitting in to training and the testing data
X = data[['housing_median_age', 'median_income']]
y = data['median_house_value']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#print("Training set shape:", X_train.shape, y_train.shape)
#print("Testing set shape:", X_test.shape, y_test.shape)
#funding the shape of the data
```

```python
#building a linear model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
#predicting the input testing data
y_pred = linear_model.predict(X_test)
```

```python
#calculating MAE MSE and r^2
MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error (MAE):", MAE)
print("Mean Squared Error (MSE):", MSE)
print("R2 Score:", r2)
```

The model has a moderate fit to the data. It  still doesn't explain the  significant part of the variance which means that  there are other  factors influencing house prices that are not captured by Housing_median_age and median_income. Also the MAE and MSE values are higher which indicates that the model predicted more errors.
